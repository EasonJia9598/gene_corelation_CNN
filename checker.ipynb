```python
def extract_numeric_part(filename):
    return int(''.join(filter(str.isdigit, filename)))

def get_files_list(data_dir, file_extension=".npy"):
        # Get a list of all files in the directory
    files = os.listdir(data_dir)

    # Filter files to include only those that match the pattern
    files = [file for file in files if file.endswith(file_extension)]

    # Sort the files based on their numeric values
    files.sort(key=extract_numeric_part)
    return files
```

```python
rates_data_dir = '/home/eeepc/Documents/random_tree/random_tree_5_genes_2024_02_10_13_38_30_n_records_200000/concatenated_data'
gene_rates_files = get_files_list(rates_data_dir, file_extension=".csv")
```

```python
substitution_rates = []
for gene_rates_file in gene_rates_files:
    gene_rates = pd.read_csv(os.path.join(rates_data_dir, gene_rates_file))
    substitution_rates.append(gene_rates)
```
